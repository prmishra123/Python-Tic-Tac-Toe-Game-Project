```python
import numpy
import random
from time import sleep
```

```python
def vt(m,n,g):
    for i in range(3):
        if(m[i][0]==g and m[i][1]==g and m[i][2]==g):
            return(0)
        elif(m[0][i]==g and m[1][i]==g and m[2][i]==g):
            return(0)
    if((m[0][0]==g and m[1][1]==g and m[2][2]==g) or (m[0][2]==g and m[1][1]==g and m[2][0]==g)):
        return(0)
    else:
        return(n)
```

```python
def s():
    n=9
    m=[[0,0,0],[0,0,0],[0,0,0]]
    numpy.mat(m)
    e=1
    while(n):
        l=[]
        for i in range(3):
            for j in range(3):
                if m[i][j]==0:
                    l.append([i,j])
        o=[['*','*','*'],['*','*','*'],['*','*','*']]
        h='abcdefghi'
        f=0
        for [i,j] in l:
            o[i][j]=h[f]
            f=f+1
        print(numpy.mat(o))
        print("enter ur choice")
        q=input()
        for i in range(3):
            for j in range(3):
                if(o[i][j]==q):
                    m[i][j]='X'
                    l.remove([i,j])
        n=n-1
        if n==0:
            print("the match has been drawn")
        sleep(2)
        print(numpy.mat(m))
        e=vt(m,e,'X')
        if e==0:
            print("congrats u have won the game")
            break
        sleep(2)
        [r,t]=random.choice(l)
        m[r][t]='O'
        print(numpy.mat(m))
        e=vt(m,e,'O')
        if e==0:
            print("you lost \n better luck next time")
            break
        n=n-1
```
